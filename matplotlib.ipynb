```python
import numpy
from matplotlib import pyplot


def show(filename):
    data = numpy.loadtxt(filename)
    pyplot.hist(data)
    pyplot.show()
```

```python
def save(filename):
    data = numpy.loadtxt(filename)
    pyplot.hist(data)
    pyplot.savefig("figure1.eps")
    
```

```python
filename = 'data/matplotilb.txt'
```

```python
show(filename)
```

```python
save(filename)
```

```python
from numpy import arange, meshgrid
from math import pi
from matplotlib import pyplot
x = arange(0,2*pi,2*pi/128)
y = arange(0,2*pi,2*pi/128)
X, Y = meshgrid(x,y)
data = numpy.random.rand(128,128)
pyplot.pcolor(X, Y, data)
```
